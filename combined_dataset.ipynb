```python
import pandas as pd
```

```python
#Import July Dataset
july = pd.read_csv("./202307-citibike-tripdata.csv")
july.head()
```

```python
#Import August Dataset
august = pd.read_csv("./202308-citibike-tripdata.csv")
august.head()
```

```python
#Import September Dataset
september = pd.read_csv("./202309-citibike-tripdata.csv")
september.head()
```

```python
#Combine All Datasets
combined_months = pd.concat([july, august, september], axis=0, ignore_index=True)
```

```python
#Export Dataset to CSV
combined_months.to_csv('combined_months.csv', index=False)
```
